# en_core_web_trf 

en_core_web_trf is a English transformer pipeline (roberta-base). It alos uses Huggingface transformer

Components: transformer, tagger, parser, ner, attribute_ruler, lemmatizer.

Download en_core_web_trf by !python -m spacy download en_core_web_trf and see more details in https://spacy.io/usage and https://huggingface.co/spacy/en_core_web_trf

```python
!python -m spacy download en_core_web_trf 
```

```python
trf = spacy.load('en_core_web_trf') # restart runtime after installation
```

Text:

Paris has parlayed her media fame into perfumes and a fashion line with her endorsement. Her fragrances have earned over $1.5 billion, and there are 50 Paris Hilton boutiques worldwide, with products including perfumes, handbags, watches and shoes.By unveiling her own luxury hotel line, entrepreneur Paris Hilton is following in the footsteps of her great-grandfather Conrad Hilton, who founded the hotel chain in the 1920s. The first of Paris Hilton's properties to open will be located in Paris, New York and Las Vegas.

```python
txt =("Paris has parlayed her media fame into perfumes and a fashion line with her endorsement. Her fragrances have earned over $1.5 billion, and there are 50 Paris Hilton boutiques worldwide, with products including perfumes, handbags, watches and shoes.By unveiling her own luxury hotel line, Paris Hilton is following in the footsteps of her great-grandfather Conrad Hilton, who founded the hotel chain in the 1920s. The first of Paris Hilton's properties to open will be located in Paris, New York and Las Vegas.")
```

```python
sm = spacy.load('en_core_web_sm') 
```

```python
doc = trf(txt)
displacy.render(doc, jupyter=True, style='ent')
```

```python
doc = sm(txt)
displacy.render(doc, jupyter=True, style='ent')
```
